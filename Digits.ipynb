```python
#Application of a simple Logistic Regression on The digits dataset that comes with scikit-learn
from sklearn.datasets import load_digits
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sns

digits = load_digits()

print('Sample Training Data')
plt.figure(figsize=(15,3))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
 plt.subplot(1, 5, index + 1)
 plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)

x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)

logisticRegr = LogisticRegression()

logisticRegr.fit(x_train, y_train)

predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
cm = metrics.confusion_matrix(y_test, predictions)
plt.figure(figsize=(10,10))
plt.title('Accuracy Score: %f\n' % score, fontsize = 15)
sns.heatmap(cm, vmin=0, vmax=9, annot=True, fmt=".2f", linewidths=1, square = True, cmap="YlGnBu");
plt.ylabel('Actual label', size = 12);
plt.xlabel('Predicted label', size = 12);

```
